# **SAMPLE 1**

```python
import os
import cv2
import numpy as np
from keras.applications.vgg16 import VGG16
from tensorflow.keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input
from sklearn.metrics.pairwise import cosine_similarity

# set the image size
img_size = (224, 224)

# load the VGG16 model
model = VGG16(weights='imagenet', include_top=False)

from google.colab import drive
drive.mount('/content/drive')


# loop through the images directory and preprocess the images
image_features_1 = []
image_paths_1 = os.listdir('/content/drive/MyDrive/flickr30k_images/sample1/')
for path in image_paths_1:
    img = image.load_img('/content/drive/MyDrive/flickr30k_images/sample1/' + path, target_size=img_size)
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    features = model.predict(x)
    features = np.squeeze(features, axis=0)  # remove batch dimension
    image_features_1.append(features)
```

```python
# convert the feature vectors to a NumPy array
image_features_1 = np.array(image_features_1)

# flatten the feature vectors
image_features_flat = image_features_1.reshape((image_features_1.shape[0], -1))

# compare the feature vectors using cosine similarity
similarity_matrix_1 = cosine_similarity(image_features_flat)

# find the indices of the most diverse images for each image
diverse_indices_1 = []
for i in range(similarity_matrix_1.shape[0]):
    indices = np.argsort(similarity_matrix_1[i])[::-1][-1:-6:-1]
    diverse_indices_1.append(indices)
```

```python
diverse_images_1 = []
os.makedirs('/content/drive/MyDrive/flickr30k_images/result/s_1', exist_ok=True)

for i in range(len(diverse_indices_1)):
    best_image_id = image_paths_1[diverse_indices_1[i][0]].split('.')[0]
    best_image_path = '/content/drive/MyDrive/flickr30k_images/sample1/' + image_paths_1[diverse_indices_1[i][0]]
    best_image = cv2.imread(best_image_path)

    if best_image is not None:
        cv2.imwrite('/content/drive/MyDrive/flickr30k_images/result/s_1/' + best_image_id + '.jpg', best_image)
        for j in range(3):
            img_id = image_paths_1[diverse_indices_1[i][j]].split('.')[0]
            img = cv2.imread('/content/drive/MyDrive/flickr30k_images/sample1/' + image_paths_1[diverse_indices_1[i][j]])
            diverse_images_1.append(img)
    else:
        print('Error: best image is empty')

    # save only the top 860 diverse images
    if i >= 860:
        break
```
