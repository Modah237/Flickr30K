```python
from transformers import T5ForConditionalGeneration, T5Tokenizer
import os

# Load the T5 model and tokenizer
model = T5ForConditionalGeneration.from_pretrained('t5-base')
tokenizer = T5Tokenizer.from_pretrained('t5-base')
```

```python


# Define the input and output file paths
input_file = 'captions.txt'
output_file = 'captions_fr.txt'

# Open the input and output files
with open(input_file, 'r') as f_in, open(output_file, 'w') as f_out:
    # Loop over the input lines and translate each one
    for line in f_in:
        # Tokenize the input text and generate the translation
        input_text = "translate English to French: " + line.strip()
        input_ids = tokenizer.encode(input_text, return_tensors='pt')
        output_ids = model.generate(input_ids=input_ids, max_length=128, num_beams=4, early_stopping=True)
        output_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)
        
        # Write the translated text to the output file
        f_out.write(output_text + '\n')

print('Translation complete.')
```

```python
# Define the input and output file paths
input_file = 'caption1.txt'
output_file = 'captions_fr_base.txt'

from transformers import T5ForConditionalGeneration, T5Tokenizer
import os
# Load the T5 model and tokenizer
model = T5ForConditionalGeneration.from_pretrained('t5-base')
tokenizer = T5Tokenizer.from_pretrained('t5-base')

# Check if GPU is available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

# Define the batch size and maximum sequence length
batch_size = 16
max_seq_len = 128

# Open the input and output files
with open(input_file, 'r') as f_in, open(output_file, 'w') as f_out:
    # Loop over the input lines and translate each one
    sentences = []
    for i, line in enumerate(f_in):
        input_text = "translate English to French: " + line.strip()
        sentences.append(input_text)
        if (i+1) % batch_size == 0 or i+1 == 3500:
            # Tokenize and generate translations for the current batch
            input_ids = tokenizer.batch_encode_plus(sentences, return_tensors='pt', padding=True, truncation=True, max_length=max_seq_len)['input_ids'].to(device)
            output_ids = model.generate(input_ids=input_ids, max_length=max_seq_len, num_beams=4, early_stopping=True)
            output_text = tokenizer.batch_decode(output_ids, skip_special_tokens=True)
            # Write the translated text to the output file
            for text in output_text:
                f_out.write(text + '\n')
            # Reset the batch
            sentences = []
            print(f"{i+1} sentences translated.")
            
print('Translation complete.')
```
